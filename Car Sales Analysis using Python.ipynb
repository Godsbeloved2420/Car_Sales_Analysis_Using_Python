```python
#IMPORTING OF RELEVANT LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#IMPORTING THE DATASET

data=pd.read_csv('C:/Users/USER/Downloads/Car_sales.csv')
```

```python
#QUESTION 3: NULL VALUE AND THEIR NUMBER IN EACH COLUMN

data.isnull().sum()
```

```python
#FILLING THE MISSING VALUES IN '__YEAR_RESALE_VALUE' WITH ITS AVERAGE

average_year_sales=data["__year_resale_value"].mean()
data["__year_resale_value"].fillna(average_year_sales, inplace=True)

#TO CHECK
data.isnull().sum()
```

```python
#DROPPING EVERY ROW WITH NULL VALUES

data= data.dropna()

#TO CHECK
data.isnull().sum()
```

```python
#QUESTION 4: CLEAN THE NAME OF THE '__YEAR_RESALE_VALUE' BY REMOVING THE LEADING UNDERSCORE

data.rename(columns={'__year_resale_value':"year_resale_value"},inplace=True)

#TO CHECK
data.columns
```

```python
#QUESTION 5A: STATISTICAL SUMMARY

data.describe()
```

```python
#QUESTION 6: MANUFACTURER WITH HIGHEST AND LOWEST AVERAGE SALES VOLUME
#AVERAGE SALES VOLUME

Average_Sales_Volume= data.groupby('Manufacturer')['Sales_in_thousands'].mean()
Average_Sales_Volume
```

```python
#HIGHEST AVERAGE SALES VOLUME

data.groupby('Manufacturer')['Sales_in_thousands'].mean().idxmax()
```

```python
#LOWEST AVERAGE SALES VOLUME
```

```python
data.groupby('Manufacturer')['Sales_in_thousands'].mean().idxmin()
```

```python
#QUESTION 7: DISTRIBUTION OF CAR PRICE IN THE DATASET

pivot1=data.pivot_table(index='Manufacturer', values= 'Price_in_thousands', aggfunc= 'sum')
pivot1.plot(kind='bar', figsize= (8,5))
plt.xlabel('Manufacturer')
plt.ylabel('Sum_of_Price_in_thousands')
plt.title('Distribution of Car Prices')
plt.show()
```

```python
#QUESTION 8: CORRELATION OF NUMERICAL VALUES IN THE DATASET

Correlate_Data= data.select_dtypes('number').corr()
Correlate_Data
```

```python
#HEAT MAP OF CORRELATION

plt.figure(figsize=(10,8))
sns.heatmap(Correlate_Data, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
```

```python
#CHANGING THE LATEST LAUNCH DATA TYPE

data.Latest_Launch= data.Latest_Launch.astype('datetime64[ns]')
```

```python
#QUESTION 9: RELATIONSHIP BETWEEN PRICE AND LATEST LAUNCH YEAR

pivot1= data.pivot_table(index= 'Latest_Launch', values= 'Price_in_thousands')
pivot1
pivot1.plot(kind= 'line',  figsize=(10,8))
plt.title('Price_in_thousands VS Latest_Launch')
plt.xlabel ('Latest_Launch')
plt.ylabel ('Price_in_thousands')
plt.grid(True)
plt.show()
```

```python
#SLICING FOR YEAR 2011 TO 2013 FOR MORE EXPLORATION

after_2010= data[data['Latest_Launch']> '01-01-2011']
after_2010
```

```python
#QUESTION 10: GROUP CARS BY MANUFACTURERS AND ANALYZE THEIR PRICE AN FUEL EFFICIENCY
#CAR GROUP AND THEIR AVERAGE FUEL EFFICIENCY

data.groupby('Manufacturer')['Fuel_efficiency'].mean()
```

```python
#CAR GROUP AND THEIR AVERAGE PRICE

data.groupby('Manufacturer')['Price_in_thousands'].mean()
```

```python
#QUESTION 11: MOST POPULAR CAR BRAND
#HIGHEST PURCHASED CAR BRANDS

df=data.groupby('Manufacturer')['Sales_in_thousands'].max().sort_values(ascending=False)
df
```

```python
#VISUALIZATION

df.plot(kind='bar', figsize=(10,8))
plt.xlabel('Manufacturer')
plt.ylabel('Sales_in_thousands')
plt.title('MOST POPULAR CAR BRANDS')
plt.show()
```

```python
#'FORD' IS THE MOST POPULAR CAR BRAND
```

```python
#QUESTION 12: CAR MODEL WITH THE HIGHEST RESALE VALUE COMPARED TO ITS INTIAL PRICE

df=data.groupby('Model')['year_resale_value'].max().sort_values(ascending=False)
df
```

```python
df.idxmax()
```

```python
#QUESTION 13: TOP 3 FUEL-EFFICIENT CARS WITH AN ENGINE SIZE ABOVE 2.5 LITRES
#fILTERING CARS WITH ENGINE SIZES ABOVE 2.5 LITRES

filtered_data= data[data['Engine_size']>2.5]
filtered_data
```

```python
#SORTING THE FILTERED DATA BY FUEL EFFICIENCY AND SELECTING THE TOP 3

Top_3_Fuel_Efficient_Cars= filtered_data.sort_values(by='Fuel_efficiency', ascending= False).head(3)
Top_3_Fuel_Efficient_Cars
```
